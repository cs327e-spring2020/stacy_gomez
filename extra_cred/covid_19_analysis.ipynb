## covid_19_analysis

#### This query lists the states and confirmed number of cases in descending order by total confirmed cases in the month of march

```python
%%bigquery
select distinct state, sum(confirmed) as confirmed
from covid_19_modeled.event e join covid_19_modeled.location l
on e.location_id = l.id
where last_update BETWEEN '2020-03-01 00:00:00' AND '2020-03-31 23:59:59'
group by state
order by confirmed desc
```

#### This query totals the number of confirmed cases and recoveries from the top 5 states for the month of march

```python
%%bigquery
select state, sum(confirmed) as confirmed, sum(recovered) as recovered
from covid_19_modeled.event e join covid_19_modeled.location l
on e.location_id = l.id
where state in ('NY', 'WA', 'IL', 'NJ', 'CA') and last_update BETWEEN '2020-03-01 00:00:00' AND '2020-03-31 23:59:59'
group by state
order by confirmed desc
```
